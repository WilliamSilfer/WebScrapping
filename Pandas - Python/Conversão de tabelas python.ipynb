```python
import pandas as pd
```

```python
file = "AlunosBolsistas.csv"
```

```python
df = pd.read_csv(file, encoding = 'ISO-8859-1', sep=';')
```

```python
df
```

```python
df["Valor Mensal da Bolsa"] = df["Valor Mensal da Bolsa"].apply(lambda x: str(x).strip('R$ '))
df["Valor Mensal da Bolsa"] = df["Valor Mensal da Bolsa"].apply(lambda x: str(x).replace(",00",""))
```

```python
df["valor Mensal da Bolsa"] = df["Valor Mensal da Bolsa"].astype('float64')
```
